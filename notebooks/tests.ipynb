```python
import psycopg
from dotenv import load_dotenv
import os
from sqlalchemy import create_engine
import pandas as pd
import numpy as np

load_dotenv()

connection = psycopg.connect(
    user=os.getenv("user"),
    password=os.getenv("password"),
    host=os.getenv("host"),
    port=os.getenv("port"),
    dbname=os.getenv("dbname")
)
cursor = connection.cursor()
print("Connected!")

conn_str = (
    f"postgresql+psycopg://{os.getenv('user')}:{os.getenv('password')}"
    f"@{os.getenv('host')}:{os.getenv('port')}/{os.getenv('dbname')}"
)
engine = create_engine(conn_str)
```

```python
# List all tables
cursor.execute("""
    SELECT table_name
    FROM information_schema.tables
    WHERE table_schema = 'public'
    ORDER BY table_name;
""")
for table in cursor.fetchall():
    print(table[0])
```

```python
# Make Dataframes
sql = "SELECT * FROM vendas"
df_vendas = pd.read_sql(sql, engine)
df_vendas.head()
```
